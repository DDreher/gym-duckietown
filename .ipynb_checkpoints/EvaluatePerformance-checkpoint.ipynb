```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(2,2,1)
df.plot(kind='line',x='Time',y=['Reward'], color='blue', figsize=(10,10), ax=ax)
ax = fig.add_subplot(2,2,2)
df.plot(kind='line',x='Time',y=['Loss'], color='red', figsize=(10,10), ax=ax)
ax = fig.add_subplot(2,2,3)
df.plot(kind='line',x='Episode',y=['Reward'], color='blue', figsize=(20,10), ax=ax)
ax = fig.add_subplot(2,2,4)
df.plot(kind='line',x='Episode',y=['Loss'], color='red', figsize=(20,10), ax=ax)


plt.show()
```
